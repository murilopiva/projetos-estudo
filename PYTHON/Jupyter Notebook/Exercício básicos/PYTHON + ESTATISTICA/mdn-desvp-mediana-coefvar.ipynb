```python
import numpy as np
import pandas as pd
import csv 

msg = ''
lListas = []

with open('arquivo.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        
        li = list(map(int, row)) # deixa a lista toda como inteiro, no arquivo estava lendo como float
        
        msg = ''
        msg += 'L' + str(line_count + 1)
        
        if len(li) != 12:
            print( 'Lista deve conter 12 elementos!')
            
        #serializa a lista para a variavel lpd
        lpd = pd.Series(li)
        
        #seta a variavel x com a contagem de cada elemento
        x = lpd.value_counts()  
        
        #verifica se a maior contagem é 1, ou seja, se nenhum está repetido
        if (max(x)) == 1:
            print('Pelo menos um elemento deve ser repetido!')
        
        #verifica se o menor elemento é menor que 1 e o maior é maior que 50
        if (np.min(li)) < 1 or (np.max(li)) > 50:
            print('Valores devem estar entre 1 e 50!')

        #verifica se a quantidade de elementos sem repetição é menor que 6
        if len(x[x==1]) < 6:
            print('Pelo menos 6 números devem ser distintos!')
            
        avg = np.average(li) # calcula a média
        mdn = np.median(li) #calcula a mediana
        var = np.std(li) ** 2 # calcula a variancia amostral
        desv = np.std(li) #calcula o desvio padrão
        coef = desv/avg #calcula o coef de variação
        md = x[x==max(x)]

        msg += ' Média: {0:.4f}'.format(avg)
        msg += ' \n - Moda: ' + str(md)
        msg += ' \n - Mediana: ' + str(np.around(mdn,4))
        msg += ' \n - Variancia: ' + str(np.around(var,4))
        msg += ' \n - Desvio Padrão: ' + str(np.around(desv,4))
        msg += ' \n - Coef. Var.: ' + str(np.around(coef,4))
        
        print (msg)
        
        lListas.append(li)
        
        line_count += 1
    
    sf = pd.Series(lListas)
    sqtd = pd.value_counts(sf)
    
    if max(sqtd) >1:
        print('Não podem haver alunos com as mesmas listas!!')
```

```python
import random 
import pandas as pd

def gerar(ini, fim, qtd):     
    i = 1
    
    arquivo = open('arquivo.txt', 'w')
    
    while i <= fim:
        res = []         
        
        for j in range(qtd): 
            rdn = random.randint(ini, fim)
            res.append(rdn)
            
            h = 0
            
            while h < len(res):
                #print (str(len(res)) + 'oi')
                
                if rdn == res[h-1]:
                    rdn = random.randint(ini, fim)
                    res[h] = rdn
                else:
                    res.append(rdn)                
                    
                h+=1
        #print (res)
                        
        lpd = pd.Series(res)
        x = lpd.value_counts()             
            
        if (max(x)) == 1:
            #print ('sem repetição')
            res.append(res[random.randint(0, 10)])
        else:
            res.append(random.randint(ini, fim))
        
        print(*res, sep = ", " )  
        #print(*res, sep = ", " , file=arquivo) 
        i += 1
        
    arquivo.close()
```

```python
gerar(1,50,11)
```
