```python
from email.message import EmailMessage
import smtplib

email = "XXXXXXXXXXXXX@gmail.com"
password = "XXXXXXXXXXXXXXXX"

msg = EmailMessage()

msg['Subject'] = 'Email using python'
msg['From'] = email
msg['To'] = email # i'm sending to myself for testing

message = "This email is generated by python!"

msg.set_content(message)

#just an example of attachment
attachments = [
    'C:\\Users\\User\\Documents\\GitHub\\projetos-estudo\\PYTHON\\Jupyter Notebook\\homer.gif'
]

for path in attachments:
    with open(path, 'rb') as file:
        data = file.read()
        name = path.split("\\")[-1]

    msg.add_attachment(data, maintype = 'application',
                        subtype = 'octet-stream',
                        filename = name)

                        
with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
    smtp.login(email,password)
    smtp.send_message(msg)

#in case of mail block the login, try to update allow less safe apps on https://myaccount.google.com/lesssecureapps
```
