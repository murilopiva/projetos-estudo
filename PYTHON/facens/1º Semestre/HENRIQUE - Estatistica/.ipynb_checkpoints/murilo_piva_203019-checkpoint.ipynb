```python
#exercício 1

def prime_number(num):
    #verifica se num é negativo e se num é float
    if num <= 0 or type(num) != int:
        print ('invalid argument')
        return           
    
    # 1 não é primo, então já sai da função
    if num == 1:
        print ('not a prime number')
        return
    
    # se o parâmetro passado for maior que o 5.5 * 10**7, então exibe mensagem
    if num > (5.5 * 10**7):
        print ('argument out of range')
        return

    div = 1    
    while div < num: # percorre loop enquanto div for menor que num
        # verificar se o o resultado de num / div é diferente dele mesmo e o resto da divisão é = 0
        # quer dizer que há outro número diferente dele que conseguiu realizar a divisão
        
        if int((num / div)) != num and (num % div) == 0:
            print ('not a prime number')
            return
        div += 1  
        
    print ('prime number')

    
#exercício 2
def sum_of_products(l1,l2):

    #percorre a lista l1 para ver se há algum elemento que não é float
    for x in l1:
        try: # tenta converter para float
            float(x)
        except: # se der erro, então é porque o elemento não é numerico
            print ('wrong number')
            return
    

    #percorre a lista l2 para ver se há algum elemento que não é float
    for x in l2:
        try:
            float(x)
        except:
            print ('wrong number')
            return        

    if len(l1) == 0 == len(l2) == 0:
        print (-1)
        
    lenl1 = len(l1)
    lenl2 = len(l2)   
    
    #descobre qual é a maior lista para o loop ser a partir dela
    if lenl1 > lenl2:
        x = lenl1
    else:
        x = lenl2

    y = 0
    prod = 0
    
    while y < x:        
        try:
            prod += l1[y] * l2[y]
        except:
            prod += 1
            
        #print (prod)
            
        y+=1
    
    return (prod)

#exercicio 3
def growth_rate(population_a,population_b):
    anos = 0
    x = population_a
    y = population_b
    
    #if not isinstance(popolation_a,int)
    if type(population_a) != int or population_a < 0 :
        print ('invalid argument')
        return
    
    if type(population_b) != int or population_b < 0 :
        print ('invalid argument')
        return    
    
    while x < y:

        x = x + (x * 0.03)
        y = y + (y * 0.015)
        
        #print ('ano:' + str(anos) + ' popA: ' + str(x) +  '                  --- popB: ' + str(y))
        anos += 1
        
    return (anos)

#exercicio 4
def count_list(l):
    #setei os key todos com 0, inicialmente
    dic={'max':0,'sum':0,'occurs':0,'mean':0,'near-mean':0,'minus':0}
    
    dic['max'] = max(l) # maior valor
    
    soma = sum(l) # soma todos os elemtos
    dic['sum'] = soma
    
    pos0 = l[0] # pega o valor o elemento 0
    dic['occurs'] = l.count(pos0) # ve quantas ocorrencias possui o valor encontrado na posição 0
    
    media = soma / len(l)# media
    dic['mean'] = media #seta a media
    
    l.sort() # ordena

    prox = l[0] #pega o valor do primeiro elemento

    for num in l:
        if (num - media)<(num-prox):
            prox = num    
    
    dic['near-mean'] = prox
    
    #soma dos valores negativos
    neg = 0
    for num in l:
        if num < 0:
            neg = neg + num
    
    dic['minus'] = neg 
    
    return (dic)
```
