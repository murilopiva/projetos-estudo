```python
#HERANÇA

class child(Parent):
    #pass
    
    def __init__ (self, name, age, grade):
        #nunca esquecer de chamar o construtor da classe pai
        super().__init__(name,age) # invoca o metodo construtor da super classe
        self.grade = grade
        
#---- fim classe

#c = child('champs',24)
#assim fica invalida a instanciação pq agora precisa de 3 argumentos
#então tem que ser assim
c = child('champs',24,10)

c.name # c é objeto da classe child mas é herança da classe Parent

#type(c)__main__.child

c.grade
```

```python
#encapsulamento

#tornar uma informação priavada
#atraves de metodos, é possível pegar o valor dos atributos privados (encapsulados)
#atraves dos metodos get e set 
#getter e setter para cada atributo encapsulado

#padrão do python é ter variável publica kkkkk e não encapsulada uaehuaheuaeh

#colando 2 underlines que rdizer que o atributo é privado

#EX:

class Account:
    def __init__(self, number):
        self.number = number
        self.__special = False
        self.__balance = 0 # inicia zerado-- conta inicia zerada
        
        
        #nesse exemplo, o atributo ballance não pode ser alterado diretamente
        #então tem os metodos
        def depostit (self, amount):
            self.__balance += amount
        
        def withdral(self,amount):
            pass         
                     
        def upgrade(self):
            self.__special = True
        
        def downgrade(self):
            if self.__balance < 0:
                raise Exception('Balance is not enough')
        
        @property
        def balance(self):
            return self.__balance
                     
        @property
        def special(self):
            return self.__special                     

c = Account(123)
c.balance = 50
c.balance

```

```python
#abstração
from abc import ABC, abstractmethod

class Employee(ABC): #isso diz que é classe abstrata
    def __init__(self,name, salary):
        self.name = name
        self.salary = salary

        #todos os empregados tem bonus
        #esta classe não sabe calcular bonus pqe depende de regra especifica de cada classe de empregado
        @abstractmethod
        def bonus():
            pass
        
#os tipos operador, manager e director extender employee
#todos os empregados tem  nome e salario
# mas por exemplo, o bonus é calculado difernete para cada tipo de empregado

class Operador(Employee):
    def bonus(self):
        return self.salary / 3
    
    def __str__(self):
        return print ('pomba')

class Manager(Employee):
    def bonus(self):
        return self.salary / 3

class Director(Employee):
    def bonus(self):
        return self.salary / 3

#e = Employee('hohn',50)# assi mda erro pq a classe bonus é abstrata

```

```python
o = Operador('hojs',1000)
o.bonus()
```
