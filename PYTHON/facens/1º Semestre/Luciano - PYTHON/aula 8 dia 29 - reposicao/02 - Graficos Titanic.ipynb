```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.who.value_counts()
```

## DistribuiÃ§Ã£o por sexo

```python
df['sex'].value_counts().plot(kind='pie', autopct='%.2f%%')
plt.show()
```

## DistribuiÃ§Ã£o pode classe

```python
sns.catplot('class', data=df, kind='count')
```

## DistribuiÃ§Ã£o por classe e sexo

```python
sns.catplot('class', data=df, kind='count', hue='sex')
```

## DistribuiÃ§Ã£o sexo e classe

```python
sns.catplot('sex', data=df, kind='count', hue='class')
```

## Sobreviventes

```python
sns.catplot('survived', data=df, kind='count').set_xticklabels(['Died', 'Survived'])
plt.show()
```

```python
sns.catplot('survived',data=df, kind='count', hue='class').set_xticklabels(['Died', 'Survived'])
plt.show()
```

```python
sns.catplot('survived', data=df, kind='count', hue='sex').set_xticklabels(['Died', 'Survived'])
plt.show()
```

## DistribuiÃ§Ã£o Adultos e CrianÃ§as

```python
sns.catplot('who', data=df, kind='count')
plt.show()
```

## Histograma de idade

```python
df.age.hist(bins=int(df.age.max()), figsize=(15, 4))
plt.show()
```

## Cidade de Embarque

```python
df.embark_town.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.show()
```

```python
sns.catplot('embark_town',data=df, hue='class', kind='count', aspect=2)
sns.catplot('embark_town',data=df, hue='alone', kind='count', aspect=2)
sns.catplot('embark_town',data=df, hue='who', kind='count', aspect=2)
plt.show()
```

```python
f, ax = plt.subplots(2, 2, figsize=(12, 8))
sns.catplot('class', 'survived', data=df, kind='point', ax=ax[0, 0])
sns.catplot('who', 'survived', data=df, kind='point', ax=ax[0, 1])
sns.catplot('embark_town', 'survived', data=df, kind='point', ax=ax[1, 0])
sns.catplot('alone', 'survived', data=df, kind='point', ax=ax[1, 1])
for i in range(2, 6): plt.close(i)
```
