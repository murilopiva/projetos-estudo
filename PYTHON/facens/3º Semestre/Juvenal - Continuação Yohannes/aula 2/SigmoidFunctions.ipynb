## 2. Erros Quadráticos Médios sobre a função Logística

2.1 Regressão Linear

```python
x = np.linspace(-15,15,100) # 100 linearly spaced numbers

y  = 3*x + 2
y2 = np.power(3*x + 2,2)

pylab.plot(x,y)
pylab.plot(x,y2)
pylab.show() # show the plot
```

2.2 Regressão logística

```python
x = np.linspace(-15,15,100) # 100 linearly spaced numbers

y  = 1 / (1 + np.exp(-x))
y2 = np.power(1 / (1 + np.exp(-x)),2)
y3 = np.power(1 - (1 / (1 + np.exp(-x))),2)

pylab.plot(x,y)
pylab.plot(x,y2)
pylab.plot(x,y3)
pylab.show() # show the plot
```

# 3. Entropia Cruzada sobre a função logística

```python
x = np.linspace(-15,15,100) # 100 linearly spaced numbers

y  = 1 / (1 + np.exp(-x))
pylab.plot(x,y)

y_true = 0
y2 = -1 * (y_true * np.log(1 / (1 + np.exp(-x))) + (1 - y_true) * np.log(1 - (1 / (1 + np.exp(-x)))))
pylab.plot(x,y2)

y_true = 1
y2 = -1 * (y_true * np.log(1 / (1 + np.exp(-x))) + (1 - y_true) * np.log(1 - (1 / (1 + np.exp(-x)))))
pylab.plot(x,y2)

pylab.show() # show the plot
```
