```python
import pandas as pd

acoes = pd.read_csv('GOOGL_Nasdaq_data.csv')
```

```python
acoes.shape
```

```python
acoes.columns
```

```python
acoes.describe()
```

```python
acoes['Close/Last'] = pd.to_numeric(acoes['Close/Last'].str.replace('$',''))
```

```python
acoes.describe()
```

### Vamos criar o rótulo, se a ação fechou acima ou abaixo do dia anterior

```python
acoes['Menos1'] = acoes['Close/Last'].shift(-1)
```

```python
acoes[['Date','Close/Last','Menos1']]
```

```python
acoes['classe'] = 'Não'
acoes['classe'][acoes['Close/Last'] - acoes['Menos1'] > 0] = 'Sim'
```

```python
acoes[['Date','Close/Last','Menos1','classe']]
```

### Vamos adicionar mais colunas

```python
acoes['Menos2'] = acoes['Close/Last'].shift(-2)
acoes['Menos3'] = acoes['Close/Last'].shift(-3)
acoes['Menos7'] = acoes['Close/Last'].shift(-7)
```

### Vamos manter apenas as colunas que interessam para classificar e excluir as linhas que ficaram com NaN no fim do conjunto

```python
acoes = acoes[['Menos1','Menos2','Menos3','Menos7','classe']][:-7]
```

```python
acoes
```

### Em seguida, vamos separar em treino e teste pra não perder a relação temporal e classificar

```python
X_treino, y_treino = acoes[['Menos1','Menos2','Menos3','Menos7']][200:], acoes[['classe']][200:]
X_teste, y_teste = acoes[['Menos1','Menos2','Menos3','Menos7']][:200], acoes[['classe']][:200]
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_treino, y_treino)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_teste)
print(accuracy_score(y_teste, y_pred))
```

```python
from sklearn.tree import export_text

r = export_text(model, X_treino.columns.tolist())
print(r)
```

### Vamos variar o parêmtro de profundidade da árvore



```python
for max_depth in [2,3,4]:
    model = DecisionTreeClassifier(max_depth = max_depth)
    model.fit(X_treino, y_treino)
    y_pred = model.predict(X_teste)
    print('Profundidade:', max_depth, '-', accuracy_score(y_teste, y_pred))
```

```python
model = DecisionTreeClassifier(max_depth = 2)
model.fit(X_treino, y_treino)
y_pred = model.predict(X_teste)
print('Profundidade:', max_depth, '-', accuracy_score(y_teste, y_pred))

r = export_text(model, X_treino.columns.tolist())
print(r)
```

## Exercício 1

E se em vez de usarmos o valor absoluto, utilizarmos a diferença? Assim a regra aprende com as diferenças. Ou ainda, se simplemente marcamos se o valor subiu ou desceu.

```python
acoes = pd.read_csv('GOOGL_Nasdaq_data.csv')
acoes.head()
```

```python
acoes.isnull().sum()
```

### Calcula a diferença da coluna "Close/Last" dos dias anteriores 

```python
acoes['Close/Last'] = pd.to_numeric(acoes['Close/Last'].str.replace('$',''))

acoes['Menos1'] = acoes['Close/Last'].shift(-1)

acoes['classe'] = 'Não'
acoes['classe'][acoes['Close/Last'] - acoes['Menos1'] > 0] = 'Sim'

acoes['Menos2'] = acoes['Close/Last'].shift(-2)
acoes['Menos3'] = acoes['Close/Last'].shift(-3)
acoes['Menos4'] = acoes['Close/Last'].shift(-4)
acoes['Menos5'] = acoes['Close/Last'].shift(-5)
acoes['Menos6'] = acoes['Close/Last'].shift(-6)
acoes['Menos7'] = acoes['Close/Last'].shift(-7)

acoes['Dif1'] = acoes['Close/Last'] - acoes['Menos1']
acoes['Dif2'] = acoes['Close/Last'] - acoes['Menos2']
acoes['Dif3'] = acoes['Close/Last'] - acoes['Menos3']
acoes['Dif4'] = acoes['Close/Last'] - acoes['Menos4']
acoes['Dif5'] = acoes['Close/Last'] - acoes['Menos5']
acoes['Dif6'] = acoes['Close/Last'] - acoes['Menos6']
acoes['Dif7'] = acoes['Close/Last'] - acoes['Menos7']
acoes[['Date','Close/Last', 'Volume','Open','High','Low','Dif1','Dif2','Dif3','Dif4','Dif5','Dif6','Dif7']]
```

### Treina e testa o Modelo considerando a diferença de valores

```python
acoesDif = acoes[['Dif1','Dif2','Dif3','Dif4','Dif5','Dif6','Dif7','classe']][:-7]

X_treino, y_treino = acoesDif[['Dif2','Dif3','Dif4','Dif5','Dif6','Dif7']][200:], acoesDif[['classe']][200:]
X_teste, y_teste = acoesDif[['Dif2','Dif3','Dif4','Dif5','Dif6','Dif7']][:200], acoesDif[['classe']][:200]

model = DecisionTreeClassifier()
model.fit(X_treino, y_treino)

y_pred = model.predict(X_teste)
print("Acurácia: ", accuracy_score(y_teste, y_pred),"\n")

r = export_text(model, X_treino.columns.tolist())
print(r)
```

### Cria colunas com valores Booleanos indicando:
   - True  - Subiu
   - False - Desceu

```python
acoes['Dif1'] = acoes['Close/Last'] > acoes['Menos1']
acoes['Dif2'] = acoes['Close/Last'] > acoes['Menos2']
acoes['Dif3'] = acoes['Close/Last'] > acoes['Menos3']
acoes['Dif4'] = acoes['Close/Last'] > acoes['Menos4']
acoes['Dif5'] = acoes['Close/Last'] > acoes['Menos5']
acoes['Dif6'] = acoes['Close/Last'] > acoes['Menos6']
acoes['Dif7'] = acoes['Close/Last'] > acoes['Menos7']
acoes[['Date','Close/Last', 'Volume','Open','High','Low','Dif1','Dif2','Dif3','Dif4','Dif5','Dif6','Dif7']]
```

### Treina e testa o Modelo considerando os valores Booleanos

```python
acoesBool = acoes[['Dif1','Dif2','Dif3','Dif4','Dif5','Dif6','Dif7','classe']][:-7]

X_treino, y_treino = acoesBool[['Dif2','Dif3','Dif4','Dif5','Dif6','Dif7']][200:], acoesBool[['classe']][200:]
X_teste, y_teste = acoesBool[['Dif2','Dif3','Dif4','Dif5','Dif6','Dif7']][:200], acoesBool[['classe']][:200]

model = DecisionTreeClassifier()
model.fit(X_treino, y_treino)

y_pred = model.predict(X_teste)
print("Acurácia: ", accuracy_score(y_teste, y_pred),"\n")

r = export_text(model, X_treino.columns.tolist())
print(r)
```

```python
for max_depth in [2,3,4,10]:
    model = DecisionTreeClassifier(max_depth = max_depth)
    model.fit(X_treino, y_treino)
    
    y_pred = model.predict(X_treino)
    print('Profundidade:', max_depth, '-', 'Acurácia no treino', accuracy_score(y_treino, y_pred))
    
    y_pred = model.predict(X_teste)
    print('Profundidade:', max_depth, '-', 'Acurácia no teste', accuracy_score(y_teste, y_pred))
    
    print()
```

```python
model = DecisionTreeClassifier(max_depth = 10)
model.fit(X_treino, y_treino)
y_pred = model.predict(X_teste)
print('Profundidade:', max_depth, '-', accuracy_score(y_teste, y_pred),"\n")

r = export_text(model, X_treino.columns.tolist())
print(r)
```

## Utilizando GridSearch para identificar os melhores parâmetros

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

param_grid = {
'criterion': ['gini', 'entropy'],
'max_depth': np.arange(1, 14),
'max_features': ['sqrt', 'log2']
}

grid = GridSearchCV(DecisionTreeClassifier(), param_grid = param_grid, cv = 5, verbose = 5, n_jobs = -1)
grid.fit(X_treino, y_treino)

print(grid.best_estimator_)
```

```python
model = grid.best_estimator_
model.fit(X_treino, y_treino)

y_pred = model.predict(X_teste)
print("Acurácia: ", accuracy_score(y_teste, y_pred),"\n")

r = export_text(model, X_treino.columns.tolist())
print(r)
```
